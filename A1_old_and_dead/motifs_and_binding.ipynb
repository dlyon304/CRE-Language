```python
import pandas as pd
import os
import glob
import numpy as np
import statistics as stats
import scipy
import math
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
import random
from src.mpra_tools.fasta_utils import write_fasta
from src.mpra_tools.predicted_occupancy import *
```

```python
activity_df = pd.read_csv("Data/activity.csv", index_col=0)
retinopathy_df = pd.read_csv("Data/retinopathy.csv", index_col=0)
L = 164
ewm = read_pwm_to_ewm("Data/PWMs/eLifeMotifs.meme")
```

```python
sequences = activity_df['sequence']

index = []
data = []

cutoff = 0.2

for label, seq in sequences.items():
    landscape = total_landscape(seq, ewm, 9)
    for loc, motifs in landscape.iterrows():
        for id in landscape.columns:
            if motifs[id]>cutoff:
                index.append((label,loc))
                data.append([id.split('_')[0],id[-1],len(ewm[id[:-2]]) ,motifs[id]])
    
    
    
    
    
```

```python
motifs_df = pd.DataFrame(
    data = data,
    index = pd.MultiIndex.from_tuples(index, names=['label','position']),
    columns=['motif','strand','length','occupancy']
)
```

```python
motifs_df.to_parquet("Data/Motifs/summarized_motifs.parquet")
```
